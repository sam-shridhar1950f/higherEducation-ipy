```python
df_test = pd.DataFrame()
```

```python
def statePrivate(state):
    privList = []
    pubList = []
    perclist = []
    cur = -1
    for val in df['State']:
        cur += 1
        if val == state:
            curPriv = df['Private'][cur]
            if curPriv == 'Yes':
                privList.append(curPriv)
            if curPriv == 'No':
                pubList.append(curPriv)
    if len(privList) > len(pubList):
        df_test[state] = pd.Series(['More Private'])
    elif len(pubList) > len(privList):
        df_test[state] = pd.Series(['More Public'])
    else:
        df_test[state] = pd.Series(['Equal'])
```

```python
for val in setList:
    statePrivate(val)
```

```python
df_test
```

```python
df6 = pd.DataFrame()
l1 = []
l2 = []
l3 = []
for val in setList:
    l1.append(val)
    l2.append(df2[val].mean())
    l3.append(df_test[val][0])
df6["State"] = l1
df6["Avg Acceptance Rate"] = l2
df6["Private or Public"] = l3
    
```
